```python
import json
prompt = "Hello, thanks for joining, I hope you are learning something new"

model_id = "amazon.titan-embed-text-v2:0"

def get_embedding(prompt, normalize, model_id=model_id):
  print(normalize," :: " ,prompt)
  body = json.dumps({
    "inputText": prompt,
    "dimensions": 1024,
    "normalize": normalize
  })

  model = bedrock_runtime_client.invoke_model(modelId=model_id, body=body, accept="application/json", contentType="application/json")
  response_body = json.loads(model.get("body").read())
  return response_body.get("embedding")
```

```python
text1 = "Python is a programming language"
text2 = "I have 2 blankets"


text1_embbeding = get_embedding(text1, normalize=True)
text2_embbeding = get_embedding(text2, normalize=True)

def cosine_similarity(vec1, vec2):
  dot_product = np.dot(vec1, vec2)
  norm_vec1 = np.linalg.norm(vec1)
  norm_vec2 = np.linalg.norm(vec2)
  similarity = dot_product / (norm_vec1 * norm_vec2)
  return similarity

cosine_similarity(text1_embbeding, text2_embbeding)
```

```python
from urllib.request import urlopen
from PIL import Image

cat_path = "./cat.jpg"
car_path = "./car.jpg"
puppy_path = "./puppy.png"

paths = [cat_path, car_path, puppy_path]
images = [Image.open(path).convert("RGBA") for path in paths]

captions = [
    "a pixelated image of a cute cat",
    "A supercar on the road \n with sunset in background",
    "a puppy sitting down",
]


import numpy as np

image_embeddings = []

for image in images:
  image_processed = clip_preprocessor(images=image, return_tensors="pt")["pixel_values"]
  image_embedding = model.get_image_features(image_processed).detach().cpu().numpy()[0]
  image_embeddings.append(image_embedding)
image_embbedings = np.array(image_embeddings)

# Embed all captions
text_embeddings = []
for caption in captions:
  inputs = clip_tokenizer(caption, return_tensors="pt")
  text_embedding = model.get_text_features(**inputs).detach().cpu().numpy()[0]
  text_embeddings.append(text_embedding)
text_embeddings = np.array(text_embeddings)
```

```python
# calculate cosine similarity between images and captions

from sklearn.metrics.pairwise import cosine_similarity

sim_matrix = cosine_similarity(text_embeddings, image_embeddings)
sim_matrix
```

```python
# create base figure

plt.figure(figsize=(10,10))
plt.imshow(sim_matrix, cmap="viridis")
# plt.xticks([])

# adjust ticks with correct labels
print(captions)
plt.yticks(range(len(captions)), captions, fontsize=12)
plt.xticks([])

for i, image in enumerate(images):
  plt.imshow(image, extent=[i-0.5, i+0.5, -1.6, -0.6], origin="lower")

for x in range(sim_matrix.shape[1]):
  for y in range(sim_matrix.shape[0]):
    plt.text(x,y,f"{sim_matrix[y,x]:.2f}", ha="center", va="center", size=30)

# Remove unnecessary spines
for sides in ["left", "top", "right", "bottom"]:
  plt.gca().spines[sides].set_visible(False)

# Resize blocks
plt.xlim(-0.5, len(captions) -0.5)
plt.ylim(len(captions) + 0.5, -2)

plt.savefig("sim_metrix.png", dpi = 300, bbox_inches="tight")
```

# **BLIP**

**üß† What is BLIP-2? (Simple explanation)**

BLIP-2 is an AI model that can:
‚úîÔ∏è Understand images
‚úîÔ∏è Understand text
‚úîÔ∏è Connect both together

Basically, BLIP-2 can look at an image and describe it, answer questions, or generate captions.

It is a Vision-Language Model (VLM).

üìå **What BLIP-2 can do**

**Examples:**

1. Image Captioning

Input: picture of a dog
Output: ‚ÄúA small puppy sitting in the grass.‚Äù

2. Visual Question Answering (VQA)

Image: cat
Question: ‚ÄúWhat color is the cat?‚Äù
Output: ‚ÄúWhite and orange.‚Äù

3. Image ‚Üí Text Embeddings

**Used for:**

Similarity search
Image-text matching
Product recognition

‚≠ê **Why BLIP-2 is special (in simple words)**

Before BLIP-2:
Vision models and language models were separate.
Connecting them required expensive training.
BLIP-2 brings a bridge called a Q-Former.

üîó BLIP-2 = Vision Encoder + Q-Former + LLM
Image ‚Üí Vision Encoder ‚Üí Q-Former ‚Üí LLM (like Flan-T5, Vicuna)


Vision Encoder ‚Üí Extracts image features

Q-Former ‚Üí Converts image features into a "language-friendly" space

LLM ‚Üí Generates text

```python
# check for windows or mac and then check for GPU/MPS or cpu
import torch

# Check CUDA (NVIDIA GPU)
if torch.cuda.is_available():
    device = "cuda"

# Check Apple Silicon GPU (M1/M2/M3)
elif torch.backends.mps.is_available():
    device = "mps"

# Default fallback
else:
    device = "cpu"

print("Using device:", device)
```

```python
from transformers import AutoProcessor, Blip2ForConditionalGeneration
import torch

blip_processor = AutoProcessor.from_pretrained("Salesforce/blip2-opt-2.7b")

model = Blip2ForConditionalGeneration.from_pretrained(
    "Salesforce/blip2-opt-2.7b",
    torch_dtype = torch.float16
)
# model.to(device)
```

```python
from PIL import Image
image = Image.open("car.jpg").convert("RGB")
image.size
```

```python
# Preprocess image

inputs = blip_processor(image, return_tensors="pt").to(device, torch.float16)
inputs["pixel_values"].shape
```

## Pre process Image

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
# convert to numpy and go from (1,3,224,224) to (224,224,3) in shape

image_inputs = inputs["pixel_values"][0].detach().cpu().numpy()
image_inputs = np.transpose(image_inputs, (1,2,0))

scaler = MinMaxScaler(feature_range=(0,255))
image_inputs = scaler.fit_transform(image_inputs.reshape(-1, image_inputs.shape[-1])).reshape(image_inputs.shape)
image_input = np.array(image_inputs, dtype=np.uint8)
Image.fromarray(image_input)
```

## Pre process text

```python
text = "Thanks for the office party"
token_ids = blip_processor(text=text, return_tensors="pt").to(device, torch.float16)
token_ids = token_ids["input_ids"][0]

# convert imput ids back to tokens
tokens = blip_processor.tokenizer.convert_ids_to_tokens(token_ids)
tokens
```

## Use case 1: Image captioning

```python
image = Image.open(car_path).convert("RGB")

inputs = blip_processor(image, return_tensors="pt").to(device, torch.float16)
inputs["pixel_values"].shape
image
```

```python
# generate image ids to be passed to the decorator (LLM)
generate_ids = model.generate(**inputs, max_new_tokens = 100)
generate_text = blip_processor.batch_decode(generate_ids, skip_special_tokens=True)[0].strip()
generate_text
#
```
